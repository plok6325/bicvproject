```python
def residual_block(ip, id):
    init = ip
    x = Conv2D(64, (3,3), activation='linear', padding='same',
                          name='res_conv_' + str(id) + '_1')(ip)
    x = BatchNormalization(axis=1,name="res_batchnorm_" + str(id) + "_1")(x)
    x = Activation('relu', name="res_activation_" + str(id) + "_1")(x)
    #x = ReflectionPadding2D()(x)
    x = Conv2D(64, (3,3), activation='linear', padding='same',
                       name='res_conv_' + str(id) + '_2')(x)
    x = BatchNormalization(axis=1, name="res_batchnorm_" + str(id) + "_2")(x)
    m = add([x,init],name='res_merge_'+str(id))
    #m = merge([x, init], mode='sum', name="res_merge_" + str(id))
        #m = Activation('relu', name="res_activation_" + str(id))(m)
    return m
```

# 一级 variable 

```python
ip = Input(shape=(80,120,3))
```

```python
x = Conv2D(64,(9,9),padding='same',activation='relu')(ip)
x = residual_block(x,1)
x = residual_block(x,2)
x = residual_block(x,3)
x = residual_block(x,4)
x = residual_block(x,5)
x = Conv2DTranspose(64,(3,3),padding='same',strides=(2,2))(x)

x = Conv2DTranspose(64,(3,3),padding='same',strides=(2,2))(x)

o = Conv2DTranspose(3,(5,5),padding='same',strides=(1,1))(x)
model = Model(inputs= ip , outputs=  o)
```

```python
opt = Adam(lr=1e-5)
```

```python
model.compile(loss='MSE',optimizer='adam')
history = model.fit(LR_images ,HR_images,batch_size= 4 ,epochs= 100  )
```

```python
lr_1e5=history.history['loss']
```

# Prediction

```python
def super_resolution(image,model):
    image = np.reshape(image,(1, image.shape[0],image.shape[1],image.shape[2]))
    exam_image_RGB= model.predict(x=image)
    exam_image_RGB = (exam_image_RGB)
    #exam_image_RGB[exam_image_RGB>1]=1
    #exam_image_RGB[exam_image_RGB<0]=0
    #exam_image_RGB *=255
    exam_image_RGB[exam_image_RGB>255]=255
    exam_image_RGB[exam_image_RGB<0]=0
    exam_image_RGB=exam_image_RGB.astype('uint8')
    exam_image_RGB = exam_image_RGB[0,:,:,:]
    return exam_image_RGB
```
