```python
def combine_images(generated_images):
    num = generated_images.shape[0]
    width = int(math.sqrt(num))
    height = int(math.ceil(float(num)/width))
    shape = generated_images.shape[2:]
    image = np.zeros((height*shape[0], width*shape[1]),
                     dtype=generated_images.dtype)
    for index, img in enumerate(generated_images):
        i = int(index/width)
        j = index % width
        image[i*shape[0]:(i+1)*shape[0], j*shape[1]:(j+1)*shape[1]] = \
            img[0, :, :]
    return image
```

# beta

```python
x = np.linspace(0,1,5000)
y11 = stats.beta.pdf(x,1,1)
y2525 = stats.beta.pdf(x,10,20)
y2426 = stats.beta.pdf(x,1,5)
y24 = stats.beta.pdf(x,5,10)
rat= y2525/y24
plt.plot(x, y11, "b",  x, y2426, "r",x,y24,"y",x, y2525, "g-",x, rat, "r--")
plt.legend([r'$\alpha=1, \beta =1 $',r'$\alpha=1 ,\beta =5 $',r'$\alpha=5 ,\beta =10 $',r'$\alpha=10 ,\beta =20 $','bf'],fontsize=14)
plt.xlabel(r'$\theta$',fontsize=13)
plt.ylabel(r'$P(\theta)$',fontsize=13,rotation=0)
# plt.savefig('beta_example')
plt.show()
plt.close()
```

```python
y2525 = stats.beta.pdf(x,25,25)
y2426 = stats.beta.pdf(x,30,20)
rat= y2525/(y2426+0.00001)
plt.figure(figsize=(15,4))
plt.subplot(121)
plt.plot( x, y2426, "r",x, y2525, "g-")
plt.plot(  (x[y2525.argmax()],x[y2525.argmax()]), (y2525.max(), 0),'g--')
plt.plot(  (x[y2426.argmax()],x[y2426.argmax()]), (y2426.max(), 0),'r--')
plt.legend([r'N:$h_N=25 , m_N=25 $',r'O:$h_O=30 ,m_O =20 $'],fontsize=16,loc=1)
plt.xlabel('BF O is %.2f \nBF N is %.2f' %(y2525.max()/y2426[y2525.argmax()], y2426.max()/y2525[y2426.argmax()]),fontsize=18)
plt.ylabel(r'$P(\theta)$',fontsize=15,rotation=0)
# plt.savefig('beta_example')

print 'BF@ 1 is %.2f BF @ 2 is %.2f' %(y2525.max()/y2426[y2525.argmax()], y2426.max()/y2525[y2426.argmax()])
```
